```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import glob
```

```python
files = glob.glob('E:\\StageIII-Colon\\CanEpiMsk\\*.npy')
saveLoc = 'E:\\StageIII-Colon\\CanEpiMsk\\CanEpiPNG\\'
```

```python
for file in files:
    temp = file.split('\\')[-1]
    nn = temp.split('.npy')[0]
    ff = np.load(file)
    plt.imsave(saveLoc+nn, ff.astype('uint8'), cmap=matplotlib.cm.gray)
    plt.close()
```
